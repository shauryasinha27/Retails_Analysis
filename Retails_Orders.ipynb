```python
df = pd.read_csv('unzipped_folder/orders.csv',na_values=['Not Available', 'unknown'])
df.head(10)
```

```python
df.columns
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df.head(5)
```

```python
df['discount']=df['list_price']*df['discount_percent']*.01
df
```

```python
df['sale_price']=df['list_price']-df['discount']
df['profit']=df['sale_price']-df['cost_price']
df
```

```python
df.dtypes
```

```python
pd.to_datetime(df['order_date'],format="%Y-%m-%d")
df
```

```python
df.drop(columns=['discount_percent','cost_price','discount'],inplace=True)
#inplace=True modifies the original DataFrame directly without creating a new one. 
df
```

```python
from sqlalchemy import create_engine

engine = create_engine("mysql+pymysql://root:SinhaGoku%4027@localhost/retail")
df.to_sql("retails_data", con=engine, if_exists="append", index=False)
#here, we use if_exists=append and not if_exists=replace as replace will create the table and give the highest values,
#eg: big int instead of int etc while append will just take the existing table and take the data types in it. So, we create
#the table schema in mysql ourselves and send this data there.
```
